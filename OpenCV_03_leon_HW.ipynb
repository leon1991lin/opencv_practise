```python
import cv2
import numpy as np
import dlib

cap = cv2.VideoCapture('./videos/homework.mp4')   # play video file
FPS = cap.get(cv2.CAP_PROP_FPS)               # frame per second
F_Count = cap.get(cv2.CAP_PROP_FRAME_COUNT)   # frame count
print(f'FPS : {FPS:.2f} ms, Frame_Count : {F_Count}')

# 獲取寬度
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))

# 獲取高度
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

print(FPS,width,height)
```

```python
# 影片拆解另存為圖片檔
i = 0
while True :
    try:
        ret, frame = cap.read()
#         frame = cv2.resize(frame, (800, 450), interpolation = cv2.INTER_LINEAR)

        if ret: 
    #         cv2.imshow('Video', frame)
            cv2.imwrite(f'./videos/frame/image{i}.jpg',frame,(cv2.IMWRITE_JPEG_QUALITY,100))
            i += 1
#             print(i)

        k = cv2.waitKey(30) & 0xff
        if k == 27:
            break
    except:
        break
        
print('end')
```

```python
#圖片合成視訊
import numpy as np
import cv2
tracker = cv2.TrackerMIL_create()

# 寫入物件的建立
# FPS = 30
# fourcc = cv2.VideoWriter_fourcc(*'XVID')
videoWrite = cv2.VideoWriter("homework_fin.mp4", -1, FPS, (width, height))

# MOG2 mask
fgbg = cv2.createBackgroundSubtractorMOG2(history = 100, detectShadows = True)

# KNN mask
bs = cv2.createBackgroundSubtractorKNN(history = 100, detectShadows=True)

# dlib 人臉辨識
detector = dlib.get_frontal_face_detector()  

bg = None

for i in range(0,955):
    if i >= 0  and i <= 70:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        videoWrite.write(img)
        cv2.waitKey(5)
        
    elif i > 70 and i <= 100:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        frame =cv2.blur(img, (3, 3))
        cv2.putText(frame, 'Blur', (100, 50), cv2.FONT_HERSHEY_TRIPLEX,2, (255, 255, 255), 2, cv2.LINE_AA)
        videoWrite.write(frame)
        cv2.waitKey(5)
        
    elif i > 100 and i <= 130:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',cv2.IMREAD_GRAYSCALE)
        sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
        sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
        sobelx = cv2.convertScaleAbs(sobelx)   # 轉回 Uint8 
        sobely = cv2.convertScaleAbs(sobely)  
        sobelxy =  cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0) 
        cv2.putText(sobelxy, 'Sobel', (100, 50), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 , 255), 2, cv2.LINE_AA)
        videoWrite.write(sobelxy)
        cv2.waitKey(10)
        
    elif i > 130 and i <= 160:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',cv2.IMREAD_GRAYSCALE)
        scharrx = cv2.Scharr(img, cv2.CV_64F,1,0)
        scharry = cv2.Scharr(img, cv2.CV_64F,0,1)
        scharrx = cv2.convertScaleAbs(scharrx)   # 轉回 Uint8 
        scharry = cv2.convertScaleAbs(scharry)  
        scharrxy =  cv2.addWeighted(scharrx, 0.5,scharry, 0.5, 0)        
        cv2.putText(scharrxy, 'Scharr', (100, 50), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 , 255), 2, cv2.LINE_AA)
        videoWrite.write(scharrxy)
        cv2.waitKey(10)
    
    elif i > 160  and i <= 200:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        videoWrite.write(img)
        cv2.waitKey(10)        
        
    elif i > 200  and i <= 220:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        b, g, r = cv2.split(img)     #分割通道
        if i % 2 == 0:
            a = cv2.merge([r,g,b])
            cv2.putText(a, 'RGBtoBGR', (100, 50), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 , 255), 2, cv2.LINE_AA)
            videoWrite.write(a)
            cv2.waitKey(4)
        else:
            b = cv2.merge([b,g,r])
            videoWrite.write(b)
            cv2.putText(b, 'RGBtoBGR', (100, 50), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 , 255), 2, cv2.LINE_AA)
            videoWrite.write(b)
            cv2.waitKey(4)
        
    elif i > 220  and i <= 240:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)  # 將 BGR圖像轉化為 HSV 圖像
        cv2.putText(hsv, 'HSV', (100, 50), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 ,255), 2, cv2.LINE_AA)             
        videoWrite.write(hsv)
        cv2.waitKey(10)
                        
    elif i > 240  and i <= 260:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)  
        H, S, V = cv2.split(hsv)
        new_hsv = cv2.merge([np.uint8(H+50) , np.uint8(S) , np.uint8(V)])
        cv2.putText(new_hsv, 'HSV H+50', (100, 50), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 ,255), 2, cv2.LINE_AA)             
        videoWrite.write(new_hsv)
        cv2.waitKey(5)

    elif i > 260  and i <= 280:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) 
        H, S, V = cv2.split(hsv)
        new_hsv = cv2.merge([np.uint8(H) , np.uint8(S+50) , np.uint8(V)])
        cv2.putText(new_hsv, 'HSV S+50', (100, 50), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 ,255), 2, cv2.LINE_AA)             
        videoWrite.write(new_hsv)
        cv2.waitKey(5)
    
    elif i > 280  and i <= 290:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)  
        H, S, V = cv2.split(hsv)
        new_hsv = cv2.merge([np.uint8(H) , np.uint8(S) , np.uint8(V+50)])
        cv2.putText(new_hsv, 'HSV V+50', (100, 50), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 ,255), 2, cv2.LINE_AA)             
        videoWrite.write(new_hsv)
        cv2.waitKey(5)

    elif i > 290  and i <= 295:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        videoWrite.write(img)
        cv2.waitKey(5) 

    elif i > 295  and i <= 360:
        frame = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        face_rects, scores, idx = detector.run(frame, 0, -.8)
        
        for i, d in enumerate(face_rects):               # 取出所有偵測的結果
            x1 = d.left(); y1 = d.top(); x2 = d.right(); y2 = d.bottom()
            text = f'{scores[i]:.2f}, ({idx[i]:0.0f})'
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 4, cv2.LINE_AA) # 以方框標示偵測的人臉
            cv2.putText(frame, text, (x1, y1), cv2.FONT_HERSHEY_DUPLEX, 0.7, (255, 255, 255), 1, cv2.LINE_AA)  # 標示分數                    
        cv2.putText(frame, 'Dlib Face Detection', (550, 550), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 , 255), 2, cv2.LINE_AA)
        videoWrite.write(frame)
        cv2.waitKey(5)    
        
    elif i > 360  and i <= 420:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',0)
        canny = cv2.Canny(img, 30, 150, apertureSize=3, L2gradient=0)
        cv2.putText(canny, 'CANNY', (900, 550), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 , 255), 2, cv2.LINE_AA)
        videoWrite.write(canny)
        cv2.waitKey(10)
        
    elif i > 420  and i <= 480:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',0)
        Laplacian = cv2.Laplacian(img, cv2.CV_64F)
        Laplacian = cv2.convertScaleAbs(Laplacian)
        cv2.putText(Laplacian, 'Laplacian', (900, 550), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 , 255), 2, cv2.LINE_AA)
        videoWrite.write(Laplacian)
        cv2.waitKey(10)
                
    elif i > 480  and i <= 550:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        fgmask = fgbg.apply(img)
        cv2.putText(fgmask, 'MOG2', (950, 550), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 , 255), 2, cv2.LINE_AA)
        videoWrite.write(fgmask)
        cv2.waitKey(10)            

    elif i > 550  and i <= 624:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        fg_mask = bs.apply(img)
        cv2.putText(fg_mask, 'KNN', (950, 550), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 , 255), 2, cv2.LINE_AA)
        videoWrite.write(fg_mask)
        cv2.waitKey(10)
        
    elif i > 624  and i <= 650:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        videoWrite.write(img)
        cv2.waitKey(10)      
        
    elif i > 650  and i <= 950:
        frame = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        if bg is None:
            bg = gray                                    # read gray which is inital image
            continue

        diff = cv2.absdiff(gray, bg)
        diff = cv2.threshold(diff, 30, 255, cv2.THRESH_BINARY)[1]  # binary 後, 只有黑和白 (0, 1)
        diff = cv2.erode(diff, None, iterations=2)
        diff = cv2.dilate(diff, None, iterations=2)
    #     diff = cv2.morphologyEx(diff,cv2.MORPH_OPEN, None, iterations=2)

        cnts, hierarchy = cv2.findContours(
            diff, 
            cv2.RETR_EXTERNAL,             # 若有重複, 只找外圈的
            cv2.CHAIN_APPROX_SIMPLE)       # 傳回特徵性座標, 如長 寬 高

        for c in cnts:
            
            if cv2.contourArea(c) < 800:    # try < 100  面積小於 800 不要 show 
                continue

            (x, y, w, h) = cv2.boundingRect(c)
            cv2.rectangle(frame, (x,y), (x+w, y+h), (0,0,255), 2)
            
        cv2.putText(frame, 'Frame Diff', (850, 550), cv2.FONT_HERSHEY_TRIPLEX,2, (255 ,255 , 255), 2, cv2.LINE_AA)
        videoWrite.write(frame)
        cv2.waitKey(10)                 

    else:
        img = cv2.imread(f'./videos/frame/image{i}.jpg',1)
        videoWrite.write(img)
        cv2.waitKey(10)
        
videoWrite.release()
print('end')
```
